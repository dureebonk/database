```python
import psycopg
```

```python
message = """ 
select * 
from language;
"""
print(message)
```

```python
conn = psycopg.connect('dbname=dvdrental user=postgres password=db1234')
cur = conn.cursor()
cur.execute(message)
rows = cur.fetchall()
```

```python
col_name = [col.name for col in cur.description]
print(col_name)
for row in rows:
    print(f'{row[0]:2d} | {row[1]} | {row[2]}')
```

```python
cur.execute("insert into language (name) values ('Klingon');")
conn.commit()
```

```python
language_list = ['Hindi', 'Tamil', 'Telugu', 'Marathi', 'Bengali']
sql_message = [f"('{language}')" for language in language_list]
print(", ".join(sql_message))
cur.execute(f"insert into language (name) values {', '.join(sql_message)};")
conn.commit()
```

```python
for language in language_list:
    cur.execute(f"delete from language where name = '{language}';")
conn.commit()
```

```python
cur.execute("update language set name = 'Korean' where name = 'Korea';")
conn.commit()
```

```python
import datetime
cur.execute(f"update language set last_update = '{datetime.datetime.now()}' where name = 'German';")
conn.commit()
```
